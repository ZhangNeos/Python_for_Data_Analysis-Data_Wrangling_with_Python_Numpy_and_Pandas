```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot(figsize=(10,10))
```

```python
df = pd.DataFrame(np.random.randn(1000,4),index=pd.date_range('1/1/2000',periods=1000), columns=list('ABCD'))
df = df.cumsum()
df.plot(figsize=(10,10))
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = pd.Series(range(len(df)))
df3.plot(x='A',figsize=(10, 10))
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a','b'])
df['b']=df['b']+np.arange(1000)
df.plot.hexbin(x='a', y='b', gridsize=25, figsize=(12, 10))
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df['b']=df['b']+np.arange(1000)
df['z']=np.random.uniform(0,3,1000)
df.plot.hexbin(x='a', y='b', C='z', reduce_C_function=np.max, gridsize=25, figsize=(12,10))
```
